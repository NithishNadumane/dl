```python
import numpy as np
import matplotlib.pyplot as plt

# Input values
x = np.linspace(-10, 10, 100)

# Activation Functions

# 1. Threshold function
def threshold(x):
    return np.where(x >= 0, 1, 0)

# 2. Sigmoid function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# 3. Tanh function
def tanh(x):
    return np.tanh(x)

# 4. ReLU function
def relu(x):
    return np.maximum(0, x)

# Compute outputs
y_threshold = threshold(x)
y_sigmoid = sigmoid(x)
y_tanh = tanh(x)
y_relu = relu(x)

# Plot graphs
plt.figure(figsize=(10,8))

plt.subplot(2,2,1)
plt.plot(x, y_threshold)
plt.title("Threshold Function")

plt.subplot(2,2,2)
plt.plot(x, y_sigmoid)
plt.title("Sigmoid Function")

plt.subplot(2,2,3)
plt.plot(x, y_tanh)
plt.title("Tanh Function")

plt.subplot(2,2,4)
plt.plot(x, y_relu)
plt.title("ReLU Function")

plt.tight_layout()
plt.show()
```
